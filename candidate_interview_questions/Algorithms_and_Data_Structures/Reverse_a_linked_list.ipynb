# Reverse a linked list

Given a singly linked-list, reverse the order of the list.

# Solution

We can reverse the linked-list with three pointers, by initiating the end of the linked list with the start, and adding to the new linked-list on the right, from the old linked-list from the left.

## Complexity

The run-time will be O(n), and the space complexity is constant, we're just moving around pointers.

```python
"""
Data Structures and helpers
"""

class Node:
    
    def __init__(self, data, next = None):
        self.data = data
        self.next = next

def print_ll(head):
    print(head.data)
    while head.next:
        head = head.next
        print(head.data)
    return None

def make_single_ll(n):
    if type(n) == int:
        l = list(range(n))
    elif type(n) == list:
        l = n
    elif type(n) == str:
        l = list(n)
    else:
        return None
    
    head = Node(l[0])
    last = head
    for i in l[1:]:
        nn = Node(i)
        last.next = nn
        last = nn
    return head
```

```python
def reverse_linked_list(head):
    p1 = head
    
    if not p1.next:
        return head
    else:
        p2 = head.next
    
    if not p2.next:
        new_head = head.next
        new_head.next = head
        head.next = None
        return new_head
    else:
        p3 = p2.next
    
    p1.next = None
    while p3.next is not None:
        p2.next = p1
        # advance
        p1 = p2
        p2 = p3
        p3 = p3.next
        
    p2.next = p1 
    p3.next = p2
    return p3

    
def run(n):
    head = make_single_ll(n)
    print("Forward: ")
    print_ll(head)
    new_head = reverse_linked_list(head)
    print("Reversed: ")
    print_ll(new_head)
    return None
```

```python
run(6)
print(" ")
run('abcdefghijklmnopqrstuvwxyz')
```
