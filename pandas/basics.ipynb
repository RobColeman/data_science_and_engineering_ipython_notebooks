```python
%matplotlib inline
#%load_ext autoreload
#%autoreload 2
%reload_ext autoreload
import numpy as np
import matplotlib.pyplot as plt
import math, sys, os
from numpy.random import randn

# and pandas
import pandas as pd

def load_delimited(path, delimiter = ","):
    f = open(path,"r+")
    X = []
    Y = []
    for l in f:
        if l == "\n":
            continue
        clean = l.replace("\n","").split(delimiter)
        X.append([float(x) for x in clean[:-1]])
        Y.append(str(clean[-1]))
    return (np.array(X),Y)
```

```python
fname = os.getcwd().replace("pandas","data") + "/iris.data"
X, Y = load_delimited(fname)
```

```python
dfX = pd.DataFrame(X, columns=['A','B','C','D'])
dfY = pd.DataFrame({'class': pd.Categorical(Y)})

df = pd.DataFrame({
    0: X[:,0],
    1: X[:,1],
    2: X[:,2],
    3: X[:,3],
    'class': pd.Categorical(Y)
    })
```

## Table Manipulation

```python
left = pd.DataFrame({'key': ['A', 'B'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['A', 'B', 'C'], 'rval': [4, 5, 6]})

pd.merge(left,right,on='key')
```

```python
pd.merge(left,right,on='key',how='outer')
```
